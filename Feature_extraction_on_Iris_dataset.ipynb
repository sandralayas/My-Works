```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = datasets.load_iris()

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])

# Define features (X) and target (y)
X = df.iloc[:, :-1]
y = df['target']

# Create a PCA model with 2 components (you can adjust this)
pca_model = PCA(n_components=2)
pca_data = pca_model.fit_transform(X)

# Visualize the PCA plot
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=y, cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Class')
plt.show()
```

```python
import pandas as pd
import plotly.express as px
from sklearn.manifold import TSNE

# Load the Iris dataset
df = px.data.iris()

# Select the features (you can adjust this)
features = df.loc[:, :'petal_width']

# Create a t-SNE model with 2 components (you can change n_components)
tsne = TSNE(n_components=2, random_state=0)
projections = tsne.fit_transform(features)

# Visualize the t-SNE plot
fig = px.scatter(projections, x=0, y=1, color=df.species, labels={'color': 'species'})
fig.show()
```
